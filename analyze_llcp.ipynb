## Reading and Exploring the XPT File

```python
import pandas as pd
import os
```

```python
# File path
file_path = "/Users/prakashsilwal/Desktop/BMI/LLCP2024.XPT"

# Verify file exists
print(f"File exists: {os.path.exists(file_path)}")
print(f"File size: {os.path.getsize(file_path) / (1024**3):.2f} GB")
```

```python
# Read the XPT file
print("Reading the XPT file (this may take a moment for large files)...")
df = pd.read_sas(file_path, format='xport')
print("Done!")
```

```python
# Display shape
print("="*70)
print("DATASET SHAPE:")
print("="*70)
print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]:,}")
print(f"Total shape: {df.shape}")
```

```python
# Display column names
print("="*70)
print("COLUMN NAMES:")
print("="*70)
columns = df.columns.tolist()
for i, col in enumerate(columns, 1):
    print(f"{i:4d}. {col}")

print(f"\nTotal columns: {len(columns)}")
```
